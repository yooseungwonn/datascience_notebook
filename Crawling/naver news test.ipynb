```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

# HTTP GET Request
req = urlopen('https://entertain.daum.net/news')
print("status:", req.getcode())

if req.getcode() == 200: # 성공
    # html 받아오기
    html = req.read()
#   print(html) # 인코딩이 깨진다
    html = html.decode("utf-8")
    print(html)
else: # 실패
    print("HTTP_ERROR")
```

```python
# html 소스 -> soup 객체로 변환
soup = BeautifulSoup(html, "html.parser") # soup 생성

# 문서 정보 확인
print("문서 제목:", soup.title.text)
```

```python
links = soup.select(".list_news")
print(len(links), "개의 기사가 있습니다")
```

```python
print(links[0]) 
```

```python
news_list = []
for link in links:
    news_list.append({
        "link": link.get("href"),
        "title": link.text.strip()
    })
    
news_list
```

```python
# 결과물 출력
print("오늘자 주요 뉴스:")
for news in news_list:
    if news.get("link").startswith("http"):
        print(f"{news.get('title')} - {news.get('link')}")
```
