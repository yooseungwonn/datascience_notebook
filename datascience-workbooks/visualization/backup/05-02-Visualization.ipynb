# Seaborn

- Matplot과 함께 사용
- 예쁜 스타일과 다양한 부가 차트 작성

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns

x = np.linspace(0, 14, 100) # 0부터 14까지 100개 구간 균등 분할
y1 = np.sin(x)
y2 = np.cos(x)
# y3 = np.tan(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="sine")
plt.plot(x, y2, label="cosine")
# plt.plot(x, y3, label='tangent')
plt.legend(loc="lower right")
plt.show()
```

```python
# 요일별 점심, 저녁, 흡연 여부
# 식사 금액과 팁을 정리한 데이터
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day',
            y="total_bill",
            hue="smoker", # 이건 나중에 보여주자 -> hue 옵션으로 데이터를 구분할 수 있다.
            data=tips)
plt.show()
```

```python
# lmplot
sns.lmplot(x="total_bill",
          y="tip",
          data=tips,
          height=7)
# lmplot 회귀선 + 플롯
# 데이터를 scatter 플롯으로 그리고, regression 그래프도 함께 그림(유효범위와 함께)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip",
           data=tips,
          hue="smoker",
           palette="Set2",
          height=7)
plt.show()
```

```python
# 연도 및 월별 항공기 승객 수 데이터셋 로드
flights = sns.load_dataset("flights")
flights.head()
```

```python
# 피봇 테이블로 변경
flights = flights.pivot("month",
                       "year",
                        "passengers")
flights.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights,
            annot=True,
            fmt="d"
           )
plt.show()
# 히트 맵(heat map)은 열을 뜻하는 히트(heat)와 지도를 뜻하는 맵(map)을 결합시킨 단어로, 색상으로
# 표현할 수 있는 다양한 정보를 일정한 이미지위에 열분포 형태의 비쥬얼한 그래픽으로 출력하는 것이 특징
```

```python
# pairplot
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
import folium
import pandas as pd
us_unemp = "../data/02. folium_US_Unemployment_Oct2012.csv"
state_data = pd.read_csv(us_unemp)
state_data.head()
```

```python
state_geo = "../data/02. folium_us-states.json"

map = folium.Map(location=[40, -90], zoom_start=4)
folium.Choropleth(geo_data=state_geo, # 지리 매핑 정보 json
                 data=state_data, # 데이터
                  columns=['State', 'Unemployment'], # 표시할 컬럼
                  key_on="feature.id",
                  fill_color="YlGn",
                  legend_name="Unemployeement Rate (%)"
                 ).add_to(map)

map
```
