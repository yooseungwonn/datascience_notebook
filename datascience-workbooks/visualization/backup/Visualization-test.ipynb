```python
# mplot3d
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 데이터 생성
t = np.linspace(0, 20, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# 3D 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z)

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 제목 설정
plt.title('3D Line Plot')

# 그래프 보여주기
plt.show()
```

```python
# 3d 산점도
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 데이터 생성
num_points = 100
np.random.seed(42)
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)

# 3D 산점도 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z, c='b', marker='o')

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 제목 설정
plt.title('3D Scatter Plot')

# 그래프 보여주기
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 데이터 생성
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# 3D 표면 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 제목 설정
plt.title('3D Surface Plot')

# 그래프 보여주기
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import art3d
from matplotlib.animation import FuncAnimation

# 데이터 생성
t = np.linspace(0, 20, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# 그래프 초기화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
line, = ax.plot3D(x, y, z, color='b')

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 제목 설정
plt.title('Interactive 3D Line Plot')

# 업데이트 함수
def update(frame):
    ax.view_init(elev=20, azim=frame)
    return [line]

# 애니메이션 생성
ani = FuncAnimation(fig, update, frames=np.arange(0, 360, 5), blit=True)

# 애니메이션을 mp4 파일로 저장하거나 화면에 보여주기
# ani.save('interactive_3d_line_plot.mp4', writer='ffmpeg')
plt.show()
```
