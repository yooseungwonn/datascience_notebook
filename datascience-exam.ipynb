## 연습문제 8
- 데이터프레임으로부터 평균 신장과 평균 체중을 구해 출력해 주십시오
- 출력 예시는 다음과 같습니다.

```
평균신장: 157.425
평균체중: 50.7
```

````python
# 여기에 코드를 작성해 주십시오
import pandas as pd

# CSV 파일에서 데이터 불러오기
df = pd.read_csv('./data/wstudents.csv')

# 평균 신장과 평균 체중 계산
h = df['height'].mean()
w = df['weights'].mean()

# 결과 출력
print(f'평균 신장: {h:.3f} ')
print(f'평균 체중: {w:.1f} ')
````
